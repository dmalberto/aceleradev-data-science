# Predicting ENEM trainees with RandomForestClassifier

The ENEM(National High School Exam, in english) is a test carried out by the National Institute of Educational Studies and Research An√≠sio Teixeira, an agency linked to the Ministry of Education of Brazil, and was created in 1998. It is used to assess the quality of high school in the country and to admission to universities.

Some students decide to take the ENEM test before finishing high school, as a test. This type of student is known as "trainee" and the IN_TREINEIRO column is the column that informs us about this characteristic.

In this challenge, we will use a binary classification model to infer whether a student is a trainer or not, with our results being 1 or 0 for trainer or not.

## General Setup

````python
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
````

````python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
````

## EDA and Feature Selection

First, let's look at the correlation table.

````python
df_train.corr()['IN_TREINEIRO'].sort_values(ascending=False).head(10)
````

Looking at the table it would make sense to choose only ```'TP_ST_CONCLUSAO'``` for training the model. However, after reading the features dictionary and using previous knowledge about the ENEM, it seems interesting to us to use the students' ages described in ```NU_IDADE``` to perform this classification as well.

Just in case, we will test each situation later.

Before it, let's check the distribution of our features and our target.

````python
df_train['IN_TREINEIRO'].value_counts()
````

````python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.countplot(df_train.query('IN_TREINEIRO == 0')['TP_ST_CONCLUSAO'], ax=ax[0])
ax[0].set_title('Distribution of high school completion situation of non trainees.')
sns.countplot(df_train.query('IN_TREINEIRO == 1')['TP_ST_CONCLUSAO'], ax=ax[1])
ax[1].set_title('Distribution of high school completion situation of trainees')
````

````python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.kdeplot(df_train.query('IN_TREINEIRO == 0')['NU_IDADE'], ax=ax[0])
ax[0].set_title('Age distribution of non-trainees')
sns.kdeplot(df_train.query('IN_TREINEIRO == 1')['NU_IDADE'], ax=ax[1])
ax[1].set_title('Age distribution of trainees')
````

**Insights:**
* We have a problem: we have a IMBALANCED DATA. Just 12% of the samples have ```IN_TREINEIRO``` value equals 1.
* All the trainees have ```TP_ST_CONCLUSAO == 3```, that is, they have not finished high school.
* Trainee students are 13 to 18 years old, while those who are not part of this group have a wider range. 


Let's fix the imbalance issue and finally train and test our model.

````python
smote = SMOTE()

X_balanced, y_balanced = smote.fit_resample(df_train[['NU_IDADE', 'TP_ST_CONCLUSAO']], df_train['IN_TREINEIRO'])
````

## Modeling

As previously mentioned, we have to check if our hypothesis of using the NU_IDADE column as a feature makes sense or not. For this, we will define a function that performs the fit, the predictor and calculates the F1 score, in addition to plotting the ROC curve.

````python
def evaluate_classifier(X, y):
    # Make a train test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=777)
        
    # Train the RF Classifier
    rf_clf = RandomForestClassifier(random_state=777,  n_estimators=300)
    rf_clf.fit(X_train, y_train)
    # Predict 'IN_TREINEIRO'
    y_pred = rf_clf.predict(X_test)
    
    # Calculate F1 Score
    print(f'F1 score:{f1_score(y_test, y_pred)}')
    
    # Plot ROC curve
    fpr, tpr, threshold = roc_curve(y_test, y_pred)
    roc_auc = auc(fpr, tpr)
                       
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
````

**Without ```NU_IDADE```:**

````python
evaluate_classifier(X_balanced[['TP_ST_CONCLUSAO']], y_balanced)
````

**With ```NU_IDADE```:**

````python
evaluate_classifier(X_balanced[['TP_ST_CONCLUSAO', 'NU_IDADE']], y_balanced)
````

Well, as you can see, using also ```NU_IDADE``` as a feature provided better results, so let's continue with this set of features.

## Generating answer to submit

As we have separate dataset to use in submitting the challenge, we can retrain our model using all the observations this time.

Also, we will adapt our answer to the required conditions: we must save the answer in a file called answer.csv with two columns: ```NU_INSCRICAO``` and ```IN_TREINEIRO```.

````python
rf_clf = RandomForestClassifier(random_state=777, n_estimators=300)
rf_clf.fit(X_balanced, y_balanced)
````

````python
y_pred = rf_clf.predict(df_test[['NU_IDADE', 'TP_ST_CONCLUSAO']])
````

````python
answer = df_test[['NU_INSCRICAO']]
answer['IN_TREINEIRO']=y_pred
````
